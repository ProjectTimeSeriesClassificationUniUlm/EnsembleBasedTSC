```python
def add_additive_white_gaussian_noise(ts, snr=10):
    # source: https://stackoverflow.com/questions/14058340/adding-noise-to-a-signal-in-python
    avg = np.mean(ts)
    avg_db = 10 * np.log10(avg)
    # Calculate noise according to [2] then convert to watts
    noise_avg_db = avg_db - snr
    noise_avg = 10 ** (noise_avg_db / 10)
    # Generate an sample of white noise
    mean_noise = 0
    noise = np.random.normal(mean_noise, np.sqrt(noise_avg), len(ts))
    # Noise up the original signal
    return ts + noise
```
